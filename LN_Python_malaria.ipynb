### **3.5 Liniendiagramm**

```python
data = {
    'Year': list(range(2000, 2021)),
    'Africa': [840000, 838000, 797000, 774000, 750000, 723000, 715000, 698000, 678000, 671000, 646000, 608000, 575000, 556000, 534000, 527000, 528000, 542000, 533000, 534000, 602000],
    'Asia': [50000, 48000, 46000, 44000, 42000, 40000, 38000, 36000, 34000, 32000, 30000, 28000, 26000, 24000, 22000, 20000, 18000, 16000, 14000, 12000, 10000],
    'Americas': [2000, 1900, 1800, 1700, 1600, 1500, 1400, 1300, 1200, 1100, 1000, 900, 800, 700, 600, 500  , 400, 300, 200, 100, 50],
    'Oceania': [500, 480, 460, 440, 420, 400, 380, 360, 340, 320, 300, 280, 260, 240, 220, 200, 180, 160, 140, 120, 100]}
df_pivot = pd.DataFrame(data).set_index('Year')
df_long = df_pivot.reset_index().melt(id_vars='Year', var_name='Region', value_name='MalariaDeaths')    
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_long, x='Year', y='MalariaDeaths', hue='Region', marker='o')
plt.title('Liniendiagramm der Malaria-Todesfälle pro Region (2000-2020)')
plt.xlabel('Jahr')
plt.ylabel('Anzahl der Todesfälle')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(df_long['Year'].unique(), rotation=45)
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

**Intepretation:**

Das Liniendiagramm zeigt einen klaren, konsistenten Abwärtstrend der Malaria-Todesfälle in allen fünf Weltregionen zwischen 2000-2020. Besonders auffällig ist, dass Afrika die absolut höchste Zahlen aufweisen, während Amerikanischen Regionen die niedrigstens Zahlen zeigen. 

Trotz des allgemeinen Rückgangs ist im Jahr 2020 in Afrika ein leichter Anstieg zu beobachten, was möglicherweise aus die COVID-19-Pandemie und unterbrochene Gesundheitsmassnahmen zurückzuführen ist.
