-	Medikamentenresistenzen in einzelnen Regionen
-	Ungleichmässige Gesundheitsversorgung zwischen Ländern und Regionen
Interpretation: Deutlich wird, dass weitere Reduktionsmassnahmen schwieriger zu erreichen sind, wenn strukturelle und biologische Grenzen eine Rolle spielen.

In den letzten Jahren des Diagramms pendeln sich die Todesfälle auf einem relativ konstanten, niedrigeren Niveau von etwa 580.000 bis 620.000 ein. Ein erneuter Rückgang bleibt aus. UNICEF und WHO führen dies unter anderem auf:
-	Finanzierungslücken in der Malariabekämpfung

Zusammenfassend zeigt das Liniendiagramm, dass die Malaria-Todesfälle in Afrika zwischen 2000 und 2020 um rund ein Drittel zurückgegangen sind.
