### **3.2.2 Histogramm der Todesfälle**

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df_long, x='MalariaDeaths', bins=30, kde=True)
plt.title('Verteilung der Malaria-Todesfälle ')
plt.xlabel('Anzahl der Todesfälle')
plt.ylabel('Häufigkeit in % der Jahre')
plt.show()
```

### **3.2.3 Boxplot pro Region**

```python
plt.figure(figsize=(8, ))
sns.boxplot(x='Region', y='MalariaDeaths', data=df_long)
plt.title('Boxplot der Malaria-Todesfälle pro Region (2000-2020)')
plt.xlabel('Region')
plt.ylabel('Anzahl der Todesfälle')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

### **3.2.4 Streuungsdiagramm (Trend über die Zeit)**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_long, x='Year', y='MalariaDeaths', hue='Region' )
plt.title('Zusammenhang zwischen Jahr und Malaria-Todesfällen')
plt.xlabel('Jahr')
plt.ylabel('Anzahl der Todesfälle')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

**Intepretation**

*Beobachtung der Streeung:*

Das Streudiagramm zeigt eine deutlich fallende Punkt Verlauf von links nach rechts, was auf einen negativen Zusammenhang zwischen dem Jahr und der Anzahl der Malaria-Todesfälle hindeutet. Mit zunehmenden Jahr liegen die Datenpunkte im Durchschnitt auf einem tieferen Niveau, was den langfristigen Rückgang der Todesfälle visuell bestätigt. 

*Streeung innerhalb  der Daten:*

Die Streeung der Datenpunkte ist nicht gleichmässig über den gesamten  Zeitraum verteilt. In den frühreren Jahren (2000-2005) ist die Streeung deutlich grösser, da insbesondere Afrika sehr hohe Todesfallzahlen aufweist, während andere Regionen bereits deutlich niederige Werte zeigen. Dies führt zu einer grossen Streeung der Punkte. 

*Veränderung der Streeung über die Zeit:*

Im weiteren Verlauf des Zeitraums nimmt die Streeung der Todesfallzahlen tendenziell ab.Die Punkte liegen ab etwa 2010 enger beieinander, was darauf hindeutet, dass sich die Malaria-Todesfälle zwischen Weltregionen zwar weiterhin untersheiden, die Unterschiede jedoch insgesamt kleiner geworden sind. 

*Regionale Unterschiede in der Streeung:*

Die farbliche Trennung nach Regionen zeigt, dass Afrika während des gesamten Zeitraums deurtlich über den anderen regionen leigt und somit den Hauptanteil der globalen Malaria-Todesfälle trägt. Die übrigen Regionen weisen eine geringere Streeung auf und bewegen sich auf einem vergleichsweise niedrigen Niveau. 

*Ausreisser und Auffälligkeiten:*

Einzelne Abweichungen vom allgemeinen Abwärtstrend sind sichtbar, insbesondere ein leichter Anstieg der Todesfälle in Afrika im Jahr 2020. Diese Abweichungen stellt jedoch keine nachhaltige Veränderung der Entwicklung dar, sondern könnte auf externe Einflüsse wie die COVID-19-Pandemie zurückzuführen sein.
