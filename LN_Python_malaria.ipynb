### 2. Daten beschaffen und laden 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("global-malaria-deaths-by-world-region.csv")
df_not_europe = df[df["Entity"] != "Europe"]
df_not_europe
```

***Beschreibung des Datensatzes***: 
- Zeilen: Jedes Jahr von 2000.2020 ist als eigener Datenpunkt aufgeführt.
- Spalten: Die verschiedenen Weltregionen sind als Spaltenüberschriften angegeben. Diese Umfassen:
    - Afrika 
    - Amerika 
    - Östliches Mittelmeer
    - Südostasien
    - Westpazifik
- Werte: In den Zellen stehen die jeweiligen Zahlen der Malaria-Todesfälle/Jahr und Region. 

**Auffälligkeiten und wichtigste Punkte:**
- Afrika weist deutlich die höchsten Zahl an Malaria-Todesfälle auf, mit über 800'000 im jahr 2000 und ca. 600'000 im jqahr 2020. 
- In den anderen Regionen sind die Todesfälle deutlich niedriger, teilweise im dreistelligen oder niedrigen vierstelligen Bereich. 
- Die Werte zeigen eine tendenzeille Abnahme der Todesfälle über die Jahre, was auf Verbesserungen in der Malariabekämpfung hindeutet.
- Europa wurde bewusst ausgelassen, da die Daten dort durchgehend bei Null liegen und somit für die Analyse keine Relevanz haben. 

#### **2.1 Malaria-Todesfälle nach Weltregion im Wide-format (2000-2020)**

## **3. Datenanalyse**

#### **3.1 Entwicklung der jährlichen Malaria-Todesfälle in Afrika (2000-2020)**

**Beschreibung des Diagramms:**
- *Trend*: 
    - Die Todesfälle nehmen über die Jahre tendenziell ab. 
    - Es gibt kleine jährliche Schwankungen. 
    - Ein leichter Anstieg ist im Jahr 2020 erkennbar. 
- *Funktion*: 
    - Das Diagramm visualisiert die Rohdaten aus dem Datensatz und macht die Entwicklung der Todesfälle über die Zeit leicht erkennbar. 

```python

jahre = list(range(2000, 2021))
todesfaelle = df_pivot['Africa'].tolist()
plt.figure(figsize=(10, 6)) 


plt.plot(jahre, todesfaelle, marker='o', linestyle='-', color='skyblue', label='Todesfälle in Afrika')

plt.xlabel('Jahr')
plt.ylabel('Anzahl der Todesfälle')
plt.title('Entwicklung der jährlichen Todesfälle in Afrika (2000-2020)')
plt.legend() 


plt.grid(True)


plt.xticks(jahre, rotation=45)


plt.show() 
```

**Interpretation Liniendiagramm:**

Das Liniendiagramm stellt die Entwicklung der jährlichen Malaria-Todesfälle in Afrika vom Jahr 2000 bis ins Jahr 2020 dar und zeigt einen insgesamt klar abnehmenden Trend, der jedoch in mehrere Phasen unterteilt werden kann.

1. **Anfangsphase (2000–2004):**

Zu Beginn des Zeitraums liegen die Malaria-Todesfälle auf einem sehr hohen Niveau von über 800.000 pro Jahr. In diesen Jahren ist kaum ein Rückgang erkennbar, was auf eine damals noch unzureichende Prävention und Behandlung hindeutet. Laut der WHO waren zu dieser Zeit wirksame Medikamente, Diagnostik und Moskitonetze in vielen Regionen Afrikas nur begrenzt verfügbar, insbesondere in ländlichen Gebieten.
Interpretation: Die hohe und stabile Zahl deutet auf strukturelle Defizite im Gesundheitswesen und eine hohe Anfälligkeit der Bevölkerung gegenüber Malaria hin.

2. **Starker Rückgang (2005–2011):**

Ab etwa 2005 erkennt man einen deutlichen und relativ schnellen Rückgang der Todesfälle. Innerhalb weniger Jahren sinkt die Zahl auf rund 650.000. Diese Phase fällt zeitlich mit der massiven Ausweitung internationaler Malariaprogramme zusammen. Die WHO  und UNICeF berichten, dass in diesem Zeitraum:
-	Millionen insektizidbehandelter Moskiotnetze verteilt wurden
-	Artemisinin-basierte Kombinationstherapien als Standardbehandlung eingeführt wurden
-	Schnelltests eine frühe Diagnose ermöglichten
Interpretation: Der starke Abwärtstrend im Diagramm lässt sich dadruch erklären, dass weniger Menschen infiziert wurden. Ausserdem wurden Erkrankte früher und wirksamer behandelt, wodurch die Sterblichkeit deutlich sank.

3. **Verlangsamter Rückgang (2012–2016):**

Zwischen 2012 und 2016 setzt sich der Rückgang fort, allerdings langsamer und mit kleineren Schwankungen. Die Todesfälle liegen nun meist zwischen 600.000 und 700.000. Laut WHO waren viele der „einfachen“ Massnahmen bereits umgesetzt. Jedoch traten neue Herausforderungen auf, z.B.:
-	Insektizid-Resistenzen bei Mücken
-	Medikamentenresistenzen in einzelnen Regionen
-	Ungleichmässige Gesundheitsversorgung zwischen Ländern und Regionen
Interpretation: Deutlich wird, dass weitere Reduktionsmassnahmen schwieriger zu erreichen sind, wenn strukturelle und biologische Grenzen eine Rolle spielen.

4. **Stagnation auf niedrigem Niveau (2017–2020):**

In den letzten Jahren des Diagramms pendeln sich die Todesfälle auf einem relativ konstanten, niedrigeren Niveau von etwa 580.000 bis 620.000 ein. Ein erneuter Rückgang bleibt aus. UNICEF und WHO führen dies unter anderem auf:
-	Finanzierungslücken in der Malariabekämpfung
-	Unterbrechungen der Gesundheitsversorgung
-	Zusätzliche Belastungen durch die COVID-19-Pandemie, die Prävention und Behandlung erschwerte.

*Interpretation:* 

Hier ist eine Art Plateau sichtbar, das darauf hindeutet, dass bestehende Massnahmen zwar wirken, aber allein nicht ausreichen, um die Todesfälle weiter stark zu senken.

Zusammenfassend zeigt das Liniendiagramm, dass die Malaria-Todesfälle in Afrika zwischen 2000 und 2020 um rund ein Drittel zurückgegangen sind. Dieser Rückgang ist vor allem auf gezielze Präventionsmassnahmen, bessere medizinicshe Behandlung und interantionale Programme zurückzuführen. Gleichzeitig macht der Verlauf deutlich, dass Fortschritte nicht linear verlaufen und ohne ausreichende Finanzierung sowie Strategien ins Stocken geraten können.

Trotz des allgemeinen Rückgangs ist im Jahr 2020 in Afrika ein leichter Anstieg zu beobachten, was möglicherweise aus die COVID-19-Pandemie und unterbrochene Gesundheitsmassnahmen zurückzuführen ist. 

```python


data = {'Year': list(range(2000, 2021)),
    'Americas':     [909, 832, 764, 726, 711, 687, 581, 503, 470, 463, 502, 464, 430, 470, 348, 414, 529, 664, 571, 509, 409],
    'Eastern Mediterranean': 
    [13700, 13900, 13200, 12200, 10600, 11500, 11500, 9800, 7200, 6900, 8800, 8000, 8100, 7700, 7900, 8300, 9500, 10200, 10900, 11500, 12300],
    'South-East Asia':  [35000, 34000, 33000, 33000, 36000, 38000, 33000, 33000, 36000, 37000, 39000, 32000, 27000, 21000, 23000, 24000, 25000, 18000, 11000, 9000, 9000],
    'Western Pacific': [6100, 5300, 4700, 5000, 5600, 4500, 4900, 3800, 3500, 4700, 3500, 3000, 3400, 4000, 3800, 2400, 2900, 3000, 3000, 2600, 32006]}

df_pivot = pd.DataFrame(data).set_index('Year')
df_long = df_pivot.reset_index().melt(id_vars='Year', var_name='Region', value_name='MalariaDeaths')    

print()

plt.figure(figsize=(12, 7))
sns.lineplot(data=df_long, x='Year', y='MalariaDeaths', hue='Region', marker='o')
plt.title('Jährliche Malaria-Todesfälle nach Weltregion (2000-2020)')
plt.xlabel('Jahr')
plt.ylabel('Anzahl der Todesfälle')
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(df_long['Year'].unique(), rotation=45) 

plt.tight_layout(rect=[0, 0, 0.85, 1]) 
plt.show()
```

**Interpretation:**

Das Diagramm zeigt die jährlichen Malaria-Todesfälle in den Weltregionen Americas, Eastern Mediterranean, South-East Asia und Western Pacific im Zeitraum von 2000 bis 2020. Afrika ist in dieser Darstellung nicht enthalten. Alle dargestellten Regionen weisen deutlich niedrigere Todesfallzahlen auf als Afrika. 
South-East Asia weist im vergleich zu den anderen Regionen die meisten Malaria-Todesfälle auf. Dafür gibt es mehrer Gründe. In Asien lebt ein sehr grosser Teil der Weltbevölkerung. Die hohe Bevölkerungszahl führt dazu, dass es mehr Ansteckungen gibt. Zudem ist Asien eine überwiegend tropische Region mit hohen Temperaturen, wodurch sich Malaria über Mücken leichter verbreiten kann. Die Lebensqualität ist in vielen Regionen Asiens nicht ideal. Sozial Ungleichheiten spielen ebenfalls eine wichtige Rolle und erklären zusammen mit der grossen Bevölkerungszahl und den günstigen klimatischen Bedingungen die vergleichsweise hohen Todeszahlen.

In Americas hat sich im betrachteten Zeitraum kaum etwas verändert. Die Linie verläuft relativ konstant, und es gibt nur sehr wenige Malaria-Todesfälle.
Im Western Pacific ist im Jahr 2019 ein starker Peak erkennbar. Die Todesfälle sind in diesem Zeitraum stark angestiegen. Ein möglicher Grund dafür könnte der Beginn der Covid-19-Pandemie sein. Während dieser Zeit wurde das Immunsystem vieler Menschen geschwächt und die Gesundheitsversorgung war stark belastet. Dadurch konnten sich sowohl Covid-19 als auch andere Krankheiten wie Malaria stärker ausbreiten.
In der Eastern Mediterranean ist zu erkennen, dass sich das Liniendiagramm über die Jahre hinweg nicht stark verändert hat. Im Jahr 2009 wurden die wenigsten Todesfälle verzeichnet, danach ist die zahl leicht angestiegen. Insgesamt bleibt die Entwicklung relativ konstant, jedoch auf einem höheren Niveau der Todesfälle als in den Americas. 
Zusammenfassend zeigt sich, dass south-East Asia im vergleich zu den anderen dargestellten Regionen die höchsten Malaria-Todesfälle aufweist, während die Americas am wenigsten betroffen sind.
