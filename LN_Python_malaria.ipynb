#### **1.2 Malaria-Todesfälle nach Weltregion im Wide-format (2000-2020)**

- Für diesen Datensatz verwenden wir das Wide-format. um die Malaria-Todesfälle verschiedener Weltregionen übersichtlicher nebeneinander darstellen und direkt miteinander vergleichen zu können.

```python
df_pivot = df_not_europe.pivot(index='Year', columns='Entity', values='malaria_deaths')
df_pivot
```

***Beschreibung des Datensatzes***: 
- Zeilen: Jedes Jahr von 2000.2020 ist als eigener Datenpunkt aufgeführt.
- Spalten: Die verschiedenen Weltregionen sind als Spaltenüberschriften angegeben. Diese Umfassen:
    - Afrika 
    - Amerika 
    - Östliches Mittelmeer
    - Südostasien
    - Westpazifik
- Werte: In den Zellen stehen die jeweiligen Zahlen der Malaria-Todesfälle/Jahr und Region. 

**Auffälligkeiten und wichtigste Punkte:**
- Afrika weist deutlich die höchsten Zahl an Malaria-Todesfälle auf, mit über 800'000 im jahr 2000 und ca. 600'000 im jqahr 2020. 
- In den anderen Regionen sind die Todesfälle deutlich niedriger, teilweise im dreistelligen oder niedrigen vierstelligen Bereich. 
- Die Werte zeigen eine tendenzeille Abnahme der Todesfälle über die Jahre, was auf Verbesserungen in der Malariabekämpfung hindeutet.
- Europa wurde bewusst ausgelassen, da die Daten dort durchgehend bei Null liegen und somit für die Analyse keine Relevanz haben.
