## **2.Rohdaten**

#### **2.1 Entwicklung der jährlichen Malaria-Todesfälle in Afrika (2000-2020)**

```python

jahre = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
todesfaelle = [840000, 838000, 797000, 774000, 750000, 723000, 715000, 698000, 678000, 671000, 646000, 608000, 575000, 556000, 534000, 527000, 528000, 542000, 533000, 534000, 602000]



plt.figure(figsize=(10, 6)) 


plt.plot(jahre, todesfaelle, marker='o', linestyle='-', color='red', label='Todesfälle in Afrika')

plt.xlabel('Jahr')
plt.ylabel('Anzahl der Todesfälle')
plt.title('Entwicklung der jährlichen Todesfälle in Afrika (2000-2020)')
plt.legend() 


plt.grid(True)


plt.xticks(jahre, rotation=45)


plt.show() 
```

**Beschreibung des Diagramms:**
- *Trend*: 
    - Die Todesfälle nehmen über die Jahre tendenziell ab. 
    - Es gibt kleine jährliche Schwankungen. 
    - Ein leichter Anstieg ist im Jahr 2020 erkennbar. 
- *Funktion*: 
    - Das Diagramm visualisiert die Rohdaten aus dem Datensatz und macht die Entwicklung der Todesfälle über die Zeit leicht erkennbar. 

### **Balkendiagramm der Datenpunkt pro Region**

```python
df_pivot = pd.DataFrame(data).set_index('Year')
df_long = df_pivot.reset_index().melt(id_vars='Year', var_name='Region', value_name='MalariaDeaths')
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(y='Region', data=df_long)
plt.title('Anzahl der Datenpunkte pro Region (Gleiche Anzahl Jahre pro Region)')
plt.xlabel('Anzahl der Datenpunkte (Jahre)')
plt.ylabel('Region')
plt.show()
```

## **Histogramm der Todesfälle**

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df_long, x='MalariaDeaths', bins=40, kde=True)
plt.title('Verteilung der Malaria-Todesfälle ')
plt.xlabel('Anzahl der Todesfälle')
plt.ylabel('Häufigkeit in % der Jahre')
plt.show()
```

### **Boxplot pro Region**

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Region', y='MalariaDeaths', data=df_long)
plt.title('Boxplot der Malaria-Todesfälle pro Region (2000-2020)')
plt.xlabel('Region')
plt.ylabel('Anzahl der Todesfälle')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```
